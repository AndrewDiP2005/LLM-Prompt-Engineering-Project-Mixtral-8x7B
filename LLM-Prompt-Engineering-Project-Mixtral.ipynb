```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig, pipeline

model_id = "mistralai/Mixtral-8x7B-Instruct-v0.1"

bnb_config = BitsAndBytes_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)

tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    quantization_config=bnb_config,
    device_map="auto"
)

pipe = pipeline("text-generation", model=model, tokenizer=tokenizer, max_new_tokens=300)

def get_llm_response(prompt_text):
    messages = [{"role": "user", "content": prompt_text}]
    output = pipe(messages, return_full_text=False)
    return output[0]['generated_text']

print("we got here")
```

# Experiment 1: Zero-shot technique

```python
prompt_1 = "Mary has 3 brothers, Joe, Jack, and James, and 2 sisters, ages 22 and 24, Anna and Katie. They are all at a cousin's house, because it's Jack's birthday. Mary's mom's only brother is also there. How many sisters does Mary's brother, James, have?"
print(prompt_1)
response_1 = get_llm_response(prompt_1)
print(response_1)
```

| Criteria              | Score | Reason                                                                                                 |
| :-------------------- | :---: | :-------------------------------------------------------------------------------------------------------------- |
| **Factual Accuracy** |  **0/2** | **Critical failure.** The final answer is factually incorrect because of a logic and counting error.         |
| Completeness          |  1/2  | The model attempts to explain its reasoning but arrives at a flawed conclusion.                                |
| Clarity & Structure   |  1/2  | The language is clear, but the reasoning is contradictory and confusing.                                        |
| Conciseness           |  2/2  | The answer is direct and to the point.                                                                        |
| **Total** | **4/8** | **A poor result due to the logic failure.** |

# Experiment 2: Chain of Thought

```python
prompt_2 = """
Mary has 3 brothers, Joe, Jack, and James, and 2 sisters, ages 22 and 24, Anna and Katie. They are all at a cousin's house, because it's Jack's birthday. Mary's mom's only brother is also there. How many sisters does Mary's brother, James, have?

Let's think step by step to solve this problem:
1. First, list all the female siblings mentioned in the family.
2. Then, list all the male siblings mentioned in the family.
3. From the perspective of one of the brothers (James), count how many sisters he has from the list of female siblings.
4. State the final answer.
"""
print(prompt_2)
response_2 = get_llm_response(prompt_2)
print(response_2)
```

| Criteria              | Score | Reason                                                                                                 |
| :-------------------- | :---: | :-------------------------------------------------------------------------------------------------------------- |
| **Factual Accuracy** |  **0/2** | **Critical failure.** The final answer is still factually incorrect. The reasoning is flawed and introduces new errors. |
| Completeness          |  1/2  | The model follows all steps but gets them wrong.                                                     |
| Clarity & Structure   |  1/2  | The structure is clear, but the logical contradictions make the content itself very confusing.                   |
| Conciseness           |  2/2  | The answer is direct and doesn't contain unnecessary text.                                                      |
| **Total** | **4/8** | **A poor result. The model failed to correctly apply logic even when given a step-by-step framework.** |
